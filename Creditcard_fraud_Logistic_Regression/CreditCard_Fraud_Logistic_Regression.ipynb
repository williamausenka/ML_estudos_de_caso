```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sns
```

```python
raw_data = pd.read_csv('creditcard.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()

```

```python
inputs = data.iloc[:,1:-1]
targets = data.iloc[:,-2:]
targets = targets.drop(['Amount'],axis=1)

```

```python
inputs_random = inputs.sample(frac = 1)
targets_random = targets.sample(frac = 1)

```

```python
target_1 = int(np.sum(targets_random))
```

```python
target_0_counter = 0
nump_inputs_random = inputs_random.to_numpy()
nump_targets_random = targets_random.to_numpy()
nump_targets_random.shape[0]
```

```python
indices_to_remove = []
```

```python
for i in range(nump_targets_random.shape[0]):
    if nump_targets_random[i] == 0:
        target_0_counter += 1
        if target_0_counter > target_1:
            indices_to_remove.append(i)           
```

```python
inputs_balanced = np.delete(nump_inputs_random, indices_to_remove, axis=0)   
targets_balanced = np.delete(nump_targets_random, indices_to_remove, axis=0) 

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs_balanced, targets_balanced, train_size=0.8 , shuffle=True, random_state=20)
```

```python
reg = LogisticRegression(max_iter=400)
oned_y_train = np.ravel(y_train)

reg.fit(x_train,oned_y_train)
```

```python
reg.score(x_train,oned_y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name = inputs.columns
summary_table20 = pd.DataFrame(columns=['Feature Name'],data = feature_name)
```

```python
summary_table20['Coefficient'] = np.transpose(reg.coef_)
```

```python
summary_table20.index = summary_table20.index + 1
summary_table20.loc[0] = ['Intercept',reg.intercept_[0]]
```

```python
summary_table20 = summary_table20.sort_values('Coefficient', ascending=False)
summary_table20.to_excel("summary_table20.xlsx")
summary_table20
```
