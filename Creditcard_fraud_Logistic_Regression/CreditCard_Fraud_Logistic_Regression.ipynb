```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('creditcard.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
inputs = data.iloc[:,1:-1]
targets = data.iloc[:,-2:]
targets = targets.drop(['Amount'],axis=1)

```

```python
target_1 = int(np.sum(targets))
```

```python
target_0_counter = 0
nump_inputs = inputs.to_numpy()
nump_targets = targets.to_numpy()
```

```python
indices_to_remove = []
```

```python
for i in range(nump_targets.shape[0]):
    if nump_targets[i] == 0:
        target_0_counter += 1
        if target_0_counter > target_1:
            indices_to_remove.append(i)           
```

```python
inputs_balanced = np.delete(nump_inputs, indices_to_remove, axis=0)   
targets_balanced = np.delete(nump_targets, indices_to_remove, axis=0) 

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs_balanced, targets_balanced, train_size=0.8 , shuffle=True, random_state=20)
```

```python
reg = LogisticRegression(max_iter=500)
oned_y_train = np.ravel(y_train)

reg.fit(x_train,oned_y_train)
```

```python
reg.score(x_train,oned_y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name = inputs.columns
summary_table25 = pd.DataFrame(columns=['Feature Name'],data = feature_name)
```

```python
summary_table25['Coefficient'] = np.transpose(reg.coef_)
```

```python
summary_table25.index = summary_table25.index + 1
summary_table25.loc[0] = ['Intercept',reg.intercept_[0]]
```

```python
summary_table25 = summary_table25.sort_values('Coefficient', ascending=False)
summary_table25.to_excel("summary_table25.xlsx")
summary_table25
```
