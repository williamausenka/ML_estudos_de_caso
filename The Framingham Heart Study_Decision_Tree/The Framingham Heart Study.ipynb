```python
inputs = data_no_null.iloc[:,1:-1]
inputs = inputs.drop(['education', 'totChol', 'BMI','heartRate'], axis=1)
target = data_no_null.iloc[:,-2:]
target = target.drop(['glucose'],axis=1)

```

```python
#
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs, target, train_size=0.8 , shuffle=True, random_state=20)

from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()
x_train_smote,y_train_smote = SMOTE.fit_resample(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree = decision_tree.fit(x_train_smote,y_train_smote)
decision_tree.score(x_train_smote,y_train_smote)
```

```python
from sklearn.metrics import accuracy_score
```

```python
dt_pred = decision_tree.predict(x_test)
```

```python
accuracy_score(y_test,dt_pred)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(y_test,dt_pred))
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_train_smote, y_train_smote, train_size=0.8 , shuffle=True, random_state=20)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler 
steps = [('under', RandomUnderSampler()), ('model',DecisionTreeClassifier())]
pipeline = Pipeline(steps=steps)
```

```python
pipeline.fit(X_train,Y_train)


```

```python
pipeline_pred = pipeline.predict(X_test)
```

```python
print(classification_report(Y_test,pipeline_pred))
```

```python
accuracy_score(Y_test,pipeline_pred)
```

```python
pipeline.score(X_train,Y_train)
```
