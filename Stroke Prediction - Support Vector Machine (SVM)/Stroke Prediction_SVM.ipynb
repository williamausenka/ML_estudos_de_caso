```python
################################################## SVM IMPLEMENTATION #############################################
```

```python
data_dum = data_noNull.copy()
```

```python
gender_column = pd.get_dummies(data_dum['gender'], drop_first = True)
gender_column
```

```python
everMarried_column = pd.get_dummies(data_dum['ever_married'], drop_first = True)
everMarried_column
```

```python
workType_column = pd.get_dummies(data_dum['work_type'], drop_first = True)
workType_column
```

```python
residenceType_column = pd.get_dummies(data_dum['Residence_type'], drop_first = True)
residenceType_column
```

```python
smokingStatus_column = pd.get_dummies(data_dum['smoking_status'], drop_first = True)
smokingStatus_column
```

```python
data_dum = data_dum.drop(['id', 'gender', 'work_type', 'Residence_type', 'smoking_status', 'ever_married'], axis=1)
data_dum
```

```python
data_dum = pd.concat([data_dum,gender_column,everMarried_column, workType_column, residenceType_column, smokingStatus_column],axis=1)
data_dum
```

```python
data_dum.columns.values
```

```python
columns_names = [
    'age', 'hypertension', 'heart_disease',
       'avg_glucose_level', 'bmi', 'Male', 'Other', 'Yes',
       'Never_worked', 'Private', 'Self-employed', 'children', 'Urban',
       'formerly smoked', 'never smoked', 'smokes','stroke'
]
```

```python
data_dum = data_dum[columns_names]
data_dum
```

```python
inputs = data_dum.iloc[:,:-1]
inputs
```

```python
target = data_dum.iloc[:,-2:]
target = target.drop(['smokes'],axis=1)
target
```

```python
data_dum.stroke.value_counts()
```

```python
#################################### BALANCED #######################################
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
x_train2,x_test2,y_train2,y_test2 = train_test_split(inputs, target, train_size=0.8 , shuffle=True, random_state=20)



from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()
x_train_smote,y_train_smote = SMOTE.fit_resample(x_train2,y_train2)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_train_smote, y_train_smote, train_size=0.8 , shuffle=True, random_state=20)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler 
steps = [('under', RandomUnderSampler()), ('model',SVC())]
pipeline = Pipeline(steps=steps)
```

```python
oned_Y_train = np.ravel(Y_train)
pipeline.fit(X_train,oned_Y_train)
```

```python
pipeline_pred = pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy_score(Y_test,pipeline_pred)
```

```python
print(classification_report(Y_test,pipeline_pred))
```

```python
confusion_matrix(Y_test, pipeline_pred)
```
