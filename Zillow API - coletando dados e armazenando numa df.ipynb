```python
# função para coletar dados 

def get_properties(df):
    page = 1
    while page < 16:
        #APICALL
        url = "https://zillow56.p.rapidapi.com/search"

        querystring = {"location":"orlando, fl","page":f'{page}'}

        headers = {
            "X-RapidAPI-Key": "xxx",
            "X-RapidAPI-Host": "xxx"
        }

        response = requests.request("GET", url, headers=headers, params=querystring)
        
        data = response.json()
        
        time.sleep(1)
        
        for prop in data['results']:
            bathrooms = prop['bathrooms']
            
            try:
                bedrooms = prop['bedrooms']
            except:
                bedrooms = 0.0
                
            city = prop['city']
            country = prop['country']
            homeStatus = prop['homeStatus']
            homeType = prop['homeType']
            isNonOwnerOccupied = prop['isNonOwnerOccupied']
            isPreforeclosureAuction = prop['isPreforeclosureAuction']
            
            try:
                livingArea = prop['livingArea']
            except:
                livingArea = 0.0  
                
            try:
                lotAreaUnit = prop['lotAreaUnit']
            except:
                lotAreaUnit = 'noUnit'
                
            try:
                lotAreaValue = prop['lotAreaValue']
            except:
                lotAreaValue = 0.0
        
            price = prop['price']
            
            try:
                rentZestimate = prop['rentZestimate']
            except:
                rentZestimate = 0.0
                
            try:
                zestimate = prop['zestimate']
            except:
                zestimate = 0.0
            
            #colocando os dados na df
            
            df  = df.append({
                    'bathrooms':bathrooms,
                    'bedrooms':bedrooms,
                    'city':city,
                    'country':country,
                    'homeStatus':homeStatus,
                    'homeType':homeType,
                    'isNonOwnerOccupied':isNonOwnerOccupied,
                    'isPreforeclosureAuction':isPreforeclosureAuction,
                    'livingArea':livingArea,
                    'lotAreaUnit':lotAreaUnit,
                    'lotAreaValue':lotAreaValue,
                    'price':price,
                    'rentZestimate':rentZestimate,
                    'zestimate':zestimate}, ignore_index=True)   
        
        page = page + 1
            
    return df
            
            
        
            
```

```python
df_data_propreties = pd.DataFrame(columns=['bathrooms','bedrooms','city','country','homeStatus','homeType','isNonOwnerOccupied',
                            'isPreforeclosureAuction','livingArea','lotAreaUnit','lotAreaValue','price',
                          'rentZestimate','zestimate'])
```

```python
df_data_propreties = get_properties(df_data_propreties)
```
