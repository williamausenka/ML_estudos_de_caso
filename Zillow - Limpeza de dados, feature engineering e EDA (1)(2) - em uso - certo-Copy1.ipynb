```python
df_2 = df.copy()
```

```python
df_2 = df_2.drop(['isNonOwnerOccupied','isPreforeclosureAuction','homeStatus','city','country'],axis=1)
```

```python
df_2
```

```python
df_sq = df_2.copy()
```

```python
def sqft_la(row):
    if row['lotAreaUnit'] == 'acres':
        return row['livingArea']*43560
    else:
        return row['livingArea']
```

```python
def sqft_lav(row):
    if row['lotAreaUnit'] == 'acres':
        return row['lotAreaValue']*43560
    else:
        return row['lotAreaValue']
```

```python
df_sq['livingArea_sqft'] = df_sq.apply(sqft_la, axis=1)
```

```python
df_sq['lotAreaValue_sqft'] = df_sq.apply(sqft_lav, axis=1)
```

```python
df_sq
```

```python
df_m = df_sq.copy()
```

```python
df_m['livingArea_m2'] = np.round(df_sq['livingArea_sqft']/10.764,1)
```

```python
df_m
```

```python
df_m['lotAreaValue_m2'] = np.round(df_sq['lotAreaValue_sqft']/10.764,1)
```

```python
df_m
```

```python
df_m2 = df_m.copy()
df_m2 = df_m2.drop(['livingArea','lotAreaUnit','lotAreaValue','lotAreaValue_sqft','livingArea_sqft'],axis=1)
```

```python
df_m2
```

```python
df_fe = df_m2.copy()
```

```python
df_fe['price/m2'] = np.round(df_fe['price']/df_fe['lotAreaValue_m2'],0)
```

```python
df_fe['bathPlusBed'] = df_fe['bathrooms'] + df_fe['bedrooms']
```

```python
df_fe
```

```python
df_fe.columns
```

```python
df_fe = df_fe[['bathrooms', 'bedrooms','bathPlusBed', 'homeType', 'lotAreaValue_m2',
               'livingArea_m2', 'rentZestimate','zestimate', 'price/m2', 'price']]
```

```python
df_fe
```
