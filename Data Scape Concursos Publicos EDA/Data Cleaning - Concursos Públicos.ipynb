```python
import pandas as pd
```

```python
df = pd.read_csv('webscraping_concursos_publicos.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
df
```

```python
df.isnull().any()
```

```python
new_df = df.copy()
```

```python
new_df = new_df.dropna()
```

```python
new_df.isnull().any()
```

```python
new_df.dtypes
```

```python
def clean_string(text):
    text = text.str.replace(' ','')
    text = text.str.replace('R$','',regex=False)
    text = text.str.replace('.','',regex=False)
    text = text.str.replace(',','.',regex=False)
    text = text.str.replace('porhora','',regex=False)
    text = text.str.replace('R4','',regex=False)
    return text
```

```python
new_df['Salario_Maximo'] = clean_string(new_df['Salario_Maximo'])
```

```python
new_df
```

```python
new_df['Salario_Maximo'] = new_df['Salario_Maximo'].astype(float)
```

```python
new_df.dtypes
```

```python
def lower_text(text):
    text = text.str.lower()
    return text
```

```python
new_df['Escolaridade'] = lower_text(new_df['Escolaridade'])
new_df['Cidade'] = lower_text(new_df['Cidade'])
new_df['Estado'] = lower_text(new_df['Estado'])
```

```python
new_df['Escolaridade'] = new_df['Escolaridade'].astype(str)
new_df['Cidade'] = new_df['Cidade'].astype(str)
new_df['Estado'] = new_df['Estado'].astype(str)
```

```python
new_df.to_csv('data_cleaning_concursos_publicos_v2.csv')
```
