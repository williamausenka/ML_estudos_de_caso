```python
from geopy.geocoders import Nominatim
import pandas as pd
import folium
import gmplot
```

```python
df = pd.read_csv('df_geolocation.csv')
```

```python
df_geo = df.copy()
```

```python
df_geo.drop(['Unnamed: 0', 'Instituto_Orgao', 'Vagas', 'Salario_Maximo', 'Escolaridade'], inplace=True, axis=1)
```

```python
df_geo
```

```python
loc = Nominatim(user_agent="GetLoc")
def get_coordinates(city_name):
    try:
        getLoc = loc.geocode(city_name)
        lat = getLoc.latitude
        long = getLoc.longitude
        return lat,long
    except:
        return None
```

```python
df_geo['Coordenadas'] = df_geo.apply(lambda x: get_coordinates(x['Cidade']), axis=1)
```

```python
df_geo
```

```python
df_geo.dropna(inplace = True)
```

```python
df_geo
```

```python
# Split single column into two columns use apply()
df_geo[['Latitude', 'Longitude']] = df_geo["Coordenadas"].apply(lambda x: pd.Series(str(x).split(",")))
```

```python
df_geo
```

```python
def clean_coordinates_lat(coordinate):
    coordinate = coordinate.replace('(', '')
    return coordinate

def clean_coordinates_long(coordinate):
    coordinate = coordinate.replace(')', '')
    return coordinate
```

```python
func2 = lambda x: clean_coordinates_lat(x)
df_geo['Latitude'] = df_geo['Latitude'].apply(func2)    


func3 = lambda x: clean_coordinates_long(x)
df_geo['Longitude'] = df_geo['Longitude'].apply(func3)  
```

```python
df_geo
```

```python
df_geo.drop(['Coordenadas'], inplace=True, axis=1)
```

```python
df_geo
```

```python
list_coor_lat=df_geo['Latitude'].values.tolist()
```

```python
list_coor_long=df_geo['Longitude'].values.tolist()
```

```python
import folium
import json

with open(r'C:\Users\willi/world-countries.json') as handle:
    country_geo = json.loads(handle.read())

for i in country_geo['features']:
    if i['properties']['name'] == 'Brazil':
        country = i
        break

m = folium.Map(location = [51.1657,10.4515],
               zoom_start=2,
               min_zoom = 2,
               max_zoom = 12)


folium.GeoJson(country,
               name='brazil').add_to(m)

folium.LayerControl().add_to(m)
```

```python
def add_coords(lat,long, mapa):
       return folium.Marker([lat,long]).add_to(mapa)
```

```python
for i in range(len(list_coor_lat)):
    add_coords(list_coor_lat[i],list_coor_long[i],m)
```
