```python
def zona(bairro):
    
    norte = ['Casa Verde', 'Cachoeirinha', 'Limão, Brasilândia', 'Freguesia do Ó', 'Jaçanã', 'Tremembé', 'Perus', 'Anhanguera', 'Pirituba', 
             'Jaraguá', 'São Domingos', 'Santana', 'Tucuruvi', 'Mandaqui', 'Vila Maria', 'Vila Guilherme', 'Vila Medeiros',
            'Jardim Maristela','Jardim Santa Terezinha','Vila Albertina','Vila Bela Vista','Vila Constança','Vila Ede','Vila Mazzei',
            'Vila Paiva','Vila Penteado','Vila Zat', 'Conjunto Residencial Novo Pacaembu']
    
    sul = ['Campo Limpo', 'Capão Redondo', 'Vila Andrade', 'Capela do Socorro', 'Cidade Dutra', 'Grajaú', 'Socorro', 'Cidade Ademar', 'Pedreira', 'Ipiranga', 'Sacomã', 'Jabaquara',
           "M'Boi Mirim", 'Jardim Angela', 'Jardim São Luís', 'Parelheiros', 'Marsilac', 'Santo Amaro', 'Campo Belo', 'Campo Grande', 'Santo Amaro', 'Moema', 'Saúde', 'Vila Mariana', 'Mirandópolis',
          'Usina Piratininga', 'Jardim da Glória', 'Alto Da Boa Vista', 'Bosque da Saúde', 'Brooklin', 'Chácara Inglesa',
          'Chácara Santo Antônio', 'Conceição','Granja Julieta','Indianópolis', 'Jardim Caravelas', 'Jardim Icaraí', 'Jardim Marajoara',
          'Jardim Prudência','Nova Piraju','Panamby','Paraíso','Parque Munhoz','Real Parque','Sumarezinho','São Judas',
          'Vila Campestre','Vila Clementino','Vila Cruzeiro','Vila Inglesa','Vila Mascote','Vila Nova Conceição','Vila Olímpia',
          'Vila Parque Jabaquara','Vila Sofia','Vila das Mercês']
    
           
    leste = ['Aricanduva', 'Carrão', 'Vila Formosa', 'Cidade Tiradentes', 'Ermelino Matarazzo', 'Ponte Rasa', 'Guaianases', 'Lajeado', 'Itaim Paulista', 
             'Vila Curuçá', 'Itaquera', 'Cidade Líder', 'José Bonifácio', 'Parque do Carmo', 'Mooca', 'Água Rasa', 'Belém', 'Brás', 'Pari', 'Tatuapé', 'Artur Alvim', 'Cangaíba', 'Penha', 'Vila Matilde', 'São Mateus', 
             'São Rafael', 'São Miguel', 'Jardim Helena', 'Vila Jacuí', 'Sapopemba', 'Vila Prudente', 'São Lucas', 'Jardim Jaú', 'Vila Nova Manchester',
            'Conjunto Residencial José Bonifácio', 'Jardim Ibitirama', 'Jardim Jau Zona Leste','Jardim Nossa Senhora do Carmo',
            'Jardim Nova Conquista','Jardim Santa Adelia', 'Maranhão','Parque São Jorge','Parque da Mooca','Vila Antonieta',
            'Vila Carrão','Vila Esperança','Vila Fernandes','Vila Pierina','Vila Regente Feijó']
    
    oeste = ['Butantã', 'Morumbi', 'Raposo Tavares', 'Rio Pequeno', 'Vila Sônia', 'Lapa', 'Barra Funda', 'Jaguara', 'Jaguaré', 
             'Perdizes', 'Vila Leopoldina', 'Pinheiros', 'Alto de Pinheiros', 'Itaim Bibi', 'Jardim Paulista', 'Pinheiros', 'Cerqueira César', 'Parque Industrial Tomas Edson',
            'Alto da Lapa','Jardim Boa Vista', 'Jardim Europa','Jardim Vazani','Jardim Íris','Jardins','Sumaré','Vila Butantã',
            'Vila Guedes','Vila São Francisco','Água Branca']
    
    centro = ['Bela Vista', 'Bom Retiro', 'Cambuci', 'Consolação', 'Liberdade', 'República', 'Santa Cecília','Sé', 'Aclimação',
             'Campos Eliseos','Vila Buarque']
    
    
    if bairro in norte:
        return 'NORTE'
    elif bairro in sul:
        return 'SUL'
    elif bairro in leste:
        return 'LESTE'
    elif bairro in oeste:
        return 'OESTE'
    elif bairro in centro:
        return 'CENTRO'
    return 'ind'
```

```python
df['zona_bairro'] = df.apply(lambda x : zona(x['bairro']), axis=1)
```
