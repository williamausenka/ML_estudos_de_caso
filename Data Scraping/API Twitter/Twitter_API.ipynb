```python
import tweepy 
import pandas as pd
```

```python
#api_keys foram apagadas por motivos de segurança
api_key="5..."
api_secret_key="J..."
access_token="1..."
access_token_secret="0..."
```

```python
auth = tweepy.OAuthHandler(api_key,api_secret_key)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth)
```

```python
n_tweets = 200
tweets = []
likes = []
time =[]

for i in tweepy.Cursor(api.user_timeline, id='CoronavirusBra1',tweet_mode='extended').items(n_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
```

```python
df = pd.DataFrame({'tweets':tweets,'likes':likes,'criado':time})
```

```python
df = df[~df.tweets.str.contains('RT')]
```
