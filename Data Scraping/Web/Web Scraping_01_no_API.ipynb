```python
from bs4 import BeautifulSoup 
import requests
```

```python
html_text = requests.get('https://www.classcentral.com/subject/data-science').text
print(html_text)
```

```python
curso = []
duracao = []
oferecido_por = []
n_reviews = []
classificacao = []
```

```python
soup = BeautifulSoup(html_text, 'lxml')
todos_cursos = soup.find_all('tr', class_='row nowrap vert-align-middle padding-vert-small border-bottom border-gray-light')

def clean_up_str(string):
    string_clean = string.text.strip()
    return string_clean
```

```python
for i in todos_cursos:
    nome_curso = i.find('span', class_='text-1 weight-semi line-tight')
    if nome_curso is not None:
        resultado = nome_curso
    else:
        resultado = -1
    curso.append(resultado)
```

```python
Nomes_Cursos = []
for nomes in curso:
    nova_string = clean_up_str(nomes)
    Nomes_Cursos.append(nova_string)
```

```python
duracao = []
for i in todos_cursos:
    duracao_curso = i.find('span', class_='hidden medium-up-inline-block small-down-text-2 text-3 large-up-margin-left-xxsmall icon-clock-charcoal icon-left-small')
    if duracao_curso is not None:
        resultado = duracao_curso
    else:
        resultado = -1
    duracao.append(resultado)

    
    

Duracao_Curso = []
for tempo in duracao:
    nova_string = clean_up_str(tempo)
    Duracao_Curso.append(nova_string)
    
```

```python
Duracao_Curso_Final = []
for i in Duracao_Curso:
    i = ' '.join(i.split())
    i = i.replace('/n', '')
    Duracao_Curso_Final.append(i)
Duracao_Curso_Final
```

```python
for i in todos_cursos:
    oferecido = i.find('a', class_='color-charcoal italic')
    if oferecido is not None:
        resultado = oferecido
    else:
        resultado = -1
    oferecido_por.append(resultado)
```

```python
Oferecido_Final = []
for i in oferecido_por:
    nova_string = clean_up_str(i)
    Oferecido_Final.append(nova_string)
```

```python
for i in todos_cursos:
    rat = i.find('span', class_='text-4 color-charcoal block')
    if rat is not None:
        resultado = rat
    else:
        resultado = -1
    classificacao.append(resultado)
```

```python
class_final = []
for i in classificacao:
    nova_string = clean_up_str(i)
    class_final.append(nova_string)
```

```python
for i in todos_cursos:
    rev = i.find('span', class_='large-down-hidden block line-tight text-4 color-gray')
    if rat is not None:
        resultado = rev
    else:
        resultado = -1
    n_reviews.append(resultado)
```

```python
reviews_final = []
for i in n_reviews:
    nova_string = clean_up_str(i)
    reviews_final.append(nova_string)
```

```python
import pandas as pd
data_science_dataframe = pd.DataFrame({'Curso':Nomes_Cursos,'classificao':class_final,'N de Reviews':reviews_final,'Plataforma':Oferecido_Final,'Durao':Duracao_Curso_Final})
```
